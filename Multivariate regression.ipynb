# Predict price of car
for example: $price = \alpha + \beta_1mileage + \beta_2age + \beta_3doors$

```python
import pandas as pd
df = pd.read_excel('http://cdn.sundog-soft.com/Udemy/DataScience/cars.xls')
```

```python
df.head()
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

X = df[['Mileage', 'Cylinder', 'Doors']]
y = df['Price']
X.head()
```

```python
x = X[['Mileage', 'Cylinder', 'Doors']].copy()
x.head()
```

```python
# normalize date (scale to range between -1 to 1)
X[['Mileage', 'Cylinder', 'Doors']] = scale.fit_transform(X[['Mileage', 'Cylinder', 'Doors']].as_matrix())
print(X.head())
est = sm.OLS(y, X).fit()
est.summary()
```

# coef ~ cylinder = 5587.4472 -> highest influence on price
